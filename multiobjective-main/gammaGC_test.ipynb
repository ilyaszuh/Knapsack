```python
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
import pandas as pd
from multitools import gamma_GC
```

```python

# Step 1: Define target correlation matrix
r = np.array([
    [1.0, 0.6, 0.3],
    [0.6, 1.0, -0.5],
    [0.3, -0.5, 1.0]
])

# Step 2: Sample from multivariate normal
n = 1000
shape, scale = 2.0, 2.0
data = gamma_GC(r,n, shape, scale)
sns.pairplot(pd.DataFrame(data, columns=["X1", "X2", "X3"]))
plt.suptitle("Gamma with Gaussian Copula", y=1.02)
plt.show()

# Optional: Check empirical correlation
print("Empirical correlation matrix function:")
print(np.corrcoef(data.T))
```
